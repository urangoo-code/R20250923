```python
from tkinter import *
class MySTATE:
    def __init__(self, root):
        self.entry_var = StringVar()
        self.entry_var.set('python')
        Label(root, text = 'Normal state').grid(row = 0, column = 0, sticky='w', 
                                                padx = 10, pady=10)
        Label(root, text = 'Disabled state').grid(row = 1, column = 0, sticky='w',
                                                  padx = 10, pady=10)
        Label(root, text = 'Readonly state').grid(row = 2, column = 0, sticky='w',
                                                  padx = 10, pady = 10)
        self.entry1 = Entry(root, textvariable=self.entry_var, width=15, state = 'normal')
        self.entry1.grid(row = 0, column = 1, padx = 10)
        self.entry2 = Entry(root, textvariable=self.entry_var, width=15, state = 'disabled')
        self.entry2.grid(row = 1, column = 1, padx = 10)
        self.entry3 = Entry(root, textvariable=self.entry_var, width=15, state = 'readonly')
        self.entry3.grid(row = 2, column = 1, padx = 10)
if __name__ == "__main__":
    root = Tk()
    myobj = MySTATE(root)
    root.mainloop()
```

```python
# Example: show and selectborderwidth option using Entry widget
from tkinter import *
from tkinter import messagebox
class LoginForm:
    def __init__(self, root):
        Label(root, text = 'Username').grid(row = 0, sticky='w', 
                                            padx = 10, pady = 10)
        Label(root, text = 'Password').grid(row = 1, sticky='w', 
                                            padx = 10, pady = 10)
        self.username_entry = Entry(root, width=15, selectborderwidth = 3)
        self.username_entry.grid(row = 0, column = 1, padx = 10)
        self.password_entry = Entry(root, width=15,show= '*')
        self.password_entry.grid(row = 1, column = 1, padx = 10)
        def login():
            info = f'Username: {self.username_entry.get()}\n'
            info += f'Password: {self.password_entry.get()}'
            messagebox.showinfo('Login', info)
        self.login_button = Button(root, text = 'Login', command = login,
                                   font = ('Tahoma',11), bg = 'LightGreen')
        self.login_button.grid(row = 2, sticky= 'we', columnspan = 2, 
                               padx = 10, pady = 10)
if __name__ == "__main__":
    root = Tk()
    login_form = LoginForm(root)
    root.title('Login Page')
    root.mainloop()
```

```python
from tkinter import *
class delete_textExample(Tk):
    def __init__(self):
        super().__init__()
        self.title('Delete Text Example')
        self.entry1= Entry(self,font = ('Arial', 12),width = 30, bd = 5)
        self.entry1.pack(side = LEFT, padx = 20, pady = 20)
        self.button1 = Button(self, text="Delete the text", 
                              command=lambda: delete_text(self, self.entry1))
        self.button1.pack(padx = 10, pady = 20)
        def delete_text(self, entry):
            entry.delete(first=0, last=9)
if __name__ == "__main__":
    root = delete_textExample()
    root.mainloop()
```

```python
from tkinter import *
class MyCursorPosition(Tk):
    def __init__(self):
        super().__init__()
        self.title('Cursor Position Changing Example')
        self.entry1= Entry(self,font = ('Arial', 12),width = 20, bd = 5)
        self.entry1.pack(side = LEFT, padx=(20, 10), pady=20)
        self.entry1.focus()
        self.entry1.insert(0,'Demonstration')
        self.entry1.icursor(0)
        self.button1 = Button(self, text="Position the cursor", 
                              bg='LightGreen', font=('Tahoma', 12),
                              command=lambda: set_position(self, self.entry1))
        self.button1.pack(padx = (0, 20), pady = 20)
        def set_position(self, entry):
            entry.icursor(3)
if __name__ == "__main__":
    root = MyCursorPosition()
    root.mainloop()
```

```python
from tkinter import *
class MyIndexSelectAdjust(Tk):
    def __init__(self):
        super().__init__()
        self.title('MyIndex and Select_adjust Example')
        font = ('Tahoma', 11)
        self.entry1= Entry(self, font = font, width = 20, bd = 5)
        self.entry1.pack(side = LEFT, padx=10, pady=10)
        self.entry1.focus()
        self.entry1.insert(0,'Demonstration')
        self.entry1.icursor(0)
        self.button1 = Button(self, text="Index", font = font,
                              command=lambda: myindex(self,self.entry1))
        self.button1.pack(side = LEFT, padx = (0, 10), pady = 10)
        self.button2 = Button(self, text="Select Adjust", font=font,
                              command=lambda: myselect_adjust(self,self.entry1))
        self.button2.pack(side = LEFT, padx = (0, 10), pady = 10)
        def myindex(self, entry):
            entry.icursor(self.entry1.index(6))
        def myselect_adjust(self, entry):
            entry.select_adjust(5)
if __name__ == "__main__":
    root = MyIndexSelectAdjust()
    root.mainloop()
```

```python
from tkinter import *

class MyValidate(Tk):
    def __init__(self):
        super().__init__()
        font = ('Tahoma', 11)
        self.number_label = Label(self, text= 'Enter the number:', font=font)
        self.number_label.pack(padx=10, pady=10)
        self.number_entry = Entry(self, font = font)
        self.number_entry.pack(padx=10)
        self.error_label = Label(self, text = '', fg='Red')
        self.error_label.pack(side=BOTTOM, padx = 10)
        self.validcmd = self.register(self.valid_callback) # V1
        self.invalidcmd = self.register(self.invalid_callback) # V2
        self.number_entry.config(validate ="key", 
                                 validatecommand =(self.validcmd, '%P'), 
                                 invalidcommand = (self.invalidcmd, '%S')) # V3
        
    def valid_callback(self, s):
        if s.isdigit():# C1
            self.error_label.config(text='')
            return True
        elif s == "": # C2
            self.error_label.config(text='')
            return True
        else: # C3
            return False
            
    def invalid_callback(self, c):
        self.error_label.config(text=(f"Invalid character '{c}'\nEnter number only!"), 
                                font = ('Verdana',10))
if __name__ == "__main__":
    root = MyValidate()
    root.title('Validation Example')
    root.mainloop()
```

# Scrollbar Widgets

```python
# Scrollbar attached to Listbox
from tkinter import *
class Scrollbar_ListBox(Tk):
    def __init__(self):
        super().__init__()
        self.scrollbar = Scrollbar(self)
        self.scrollbar.pack(side=RIGHT, fill="y")
        self.listbox = Listbox(self)
        self.listbox.config(yscrollcommand=self.scrollbar.set)
        for loop in range(50):
            self.listbox.insert(END, str(loop))
        self.listbox.pack(side="left", fill=BOTH, padx=10, pady=10)
        self.scrollbar.config(command=self.listbox.yview)
if __name__ == '__main__':
    root = Scrollbar_ListBox()
    root.mainloop()
```

```python
# Scrollbar attached to Text
from tkinter import *
class Scrollbar_Text(Tk):
    def __init__(self):
        super().__init__()
        self.vscrollbar = Scrollbar(self)
        self.vscrollbar.pack(side=RIGHT, fill=Y)
        self.hscrollbar = Scrollbar(self,orient = HORIZONTAL)
        self.hscrollbar.pack(side = BOTTOM, fill = X)
        self.textbox = Text(self, width = 600, height = 600, 
                            yscrollcommand=self.vscrollbar.set, 
                            canvascommand=self.hscrollbar.set, wrap = NONE)
        self.textbox.pack(expand = 0, fill=BOTH, padx=10, pady=10)
        for loop in range(26):
            self.textbox.insert(END, str(loop) + '\t')
        for loop in range(50):
            self.textbox.insert(END, str(loop) + '\n')
        self.hscrollbar.config(command=self.textbox.xview)
        self.vscrollbar.config(command=self.textbox.yview)

if __name__ == '__main__':
    root = Scrollbar_Text()
    root.geometry('300x200')
    root.mainloop()
```

```python
# Scrollbar attached to Canvas
from tkinter import *
class Scrollbar_Canvas(Tk):
    def __init__(self):
        super().__init__()
        canvas = Canvas(self, width=150, height=50)
        canvas.create_oval(20, 20, 80, 80, fill="red")
        canvas.create_oval(200, 200, 280, 280, fill="blue")
        canvas.grid(row=0, column=0)
        xscroll = Scrollbar(self, orient="horizontal", 
                            command=canvas.xview)
        xscroll.grid(row=1, column=0, sticky=EW)
        yscroll = Scrollbar(self, command=canvas.yview)
        yscroll.grid(row=0, column=1, sticky=NS)
        canvas.configure(scrollregion=canvas.bbox("all"))
if __name__ == '__main__':
    root = Scrollbar_Canvas()
    root.geometry('200x150')
    root.mainloop()
```

```python
# Scrollbar attached to Entry
from tkinter import *
class Scrollbar_Entry(Tk):
    def __init__(self):
        super().__init__()
        self.hscrollbar = Scrollbar(self,orient = HORIZONTAL)
        self.hscrollbar.pack(side = BOTTOM,fill = X)
        self.entry1 = Entry(self,xscrollcommand=self.hscrollbar.set)
        self.entry1.pack(expand = 0, fill=BOTH)
        for loop in range(26):
            self.entry1.insert(END, str(loop) + '\t')
        self.hscrollbar.config(command=self.entry1.xview)
if __name__ == '__main__':
    root = Scrollbar_Entry()
    root.geometry('300x100')
    root.mainloop() 
```

```python
# Display string values and perform callback in Spinbox
from tkinter import *
root = Tk()
root.geometry('300x150')
a1 = StringVar()

def callback():
    root.configure(bg = a1.get())
values = ['Red','Green','Blue','Violet','Magenta','Yellow']
spinbox1 = Spinbox(font = ('Tahoma', 11), textvariable = a1, 
                   command = callback, values = values)
spinbox1.pack()
root.mainloop()
```
